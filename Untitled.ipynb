```python
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv("train.csv")

# Show first 5 rows
df.head()
```

```python
# Check missing values
print(df.isnull().sum())

# Fill missing 'Age' with median value
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill 'Embarked' with most common value
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop 'Cabin' because it has too many missing values
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visual check
sns.boxplot(x=df['Age'])
plt.show()

# Optional: remove ages above 80 if you decide they're outliers
df = df[df['Age'] <= 80]
```

```python
sns.countplot(x='Sex', data=df)
plt.title("Count of Passengers by Sex")
plt.show()

sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
df['Age'].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

df['Fare'].hist(bins=20)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
import numpy as np

# Get only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Plot correlation heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survival')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survival')
plt.show()
```

## Summary of Insights

- Most passengers in 3rd class did not survive.
- Women had a much higher survival rate than men.
- Younger passengers had a slightly higher chance of survival.
- Fare and Pclass are negatively correlated (richer people in higher classes).
